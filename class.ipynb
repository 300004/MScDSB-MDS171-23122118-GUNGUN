```python
import random

class Restaurant:
    def __init__(self):
        self.menu = {
            "rice": 30,
            "idli": 10,
        }
        self.orders = {}  # orderid: {set of order details}

    def printMenu(self):
        print("The current options in the menu are:")
        print("*" * 30)
        print(self.menu)
        print("*" * 30)

    def generateorderid(self):
        orderid = ""
        for i in range(5):
            orderid += str(random.randint(i,10))
        return orderid

    def collectorder(self):
        self.printMenu()  # print menu
        phone = input("Enter customer phone number: ")
        order = {}
        while True:
            item = input("Enter the item you want to order (or 'done' to finish): ").strip().title()
            if item == 'Done':
                break
            qty = int(input("Enter the quantity of the item: "))
            if item in self.menu:
                order[item] = qty
            else:
                print(f"{item} is not in the menu. Please choose an item from the menu.")

            orderid = self.generateorderid()
        self.orders[orderid] = {
            "Phone": phone,
            "Order": order,
        }
        print("Order placed successfully!")

    def billing(self):
        total_amount = 0
        for orderid in self.orders.keys():
            order = self.orders[orderid]["Order"]
            for item, qty in order.items():
                price = self.menu[item]
                total_amount += qty * price

        print("Billing Details:")
        print("*" * 30)
        for orderid, details in self.orders.items():
            phone = details["Phone"]
            order = details["Order"]
            print(f"Order ID: {orderid}")
            print(f"Phone Number: {phone}")
            print("Items ordered:")
            for item, qty in order.items():
                print(f"{item}: {qty} x ${self.menu[item]} = ${qty * self.menu[item]}")
            print("*" * 30)
        print(f"Total Amount: ${total_amount}")


cftb = Restaurant()
cftb.collectorder()
cftb.billing()
```
