```python

class SportMart:
    def __init__(self):
        self.inventory={}
        self.orders={}

    def customer_order(self,OrderID,Item,quantity):
        temp={
            "OrderID":OrderID,
            "Item":Item,
            "Quantity":quantity
        }
        self.orders[OrderID]=temp


    def create_new_order(self):
        OrderId=input("Enter new orderid")
        Item=input("Enter the item")
        quantity=input("enter the quantity")
        self.customer_order(OrderId,Item,quantity)
        for orderid in self.inventory:
            if self.inventory[orderid]["ProductName"]==Item:
                orderid1=orderid
        self.update_inventory(orderid1,Item)
        
    def createInventory(self,ProductID,ProductName,Quantity,Price):
        temp={
            "ProductID":ProductID,
            "ProductName":ProductName,
            "Quantity":Quantity,
            "Price":Price
        }
        self.inventory[ProductID]=temp


    def update_inventory(self,item,qty):
        # item = product id
        self.inventory[item]["Quantity"] = int(self.inventory[item]["Quantity"]) - int(qty)
                    
    def printOrders(self):
        print(self.orders)


    def printInventory(self):
        print(self.inventory)

    def printUpdatedInventory(self):
        print("new updated inventory is")
        print(self.inventory)


trinity=SportMart()
orders=open("orders.csv","r")
orders_header=orders.readline()
orders_orders=orders.readlines()
for item in orders_orders:
    temp=item.strip().split(',')
    trinity.customer_order(temp[0],temp[1],temp[2])

inventory=open("inv.csv","r")
inventory_header=inventory.readline()
inventory_inventory=inventory.readlines()
for items in inventory_inventory:
    temp=items.strip().split(',')
    trinity.createInventory(temp[0],temp[1],temp[2],temp[3])

trinity.create_new_order()

trinity.printOrders()
trinity.printInventory()
trinity.printUpdatedInventory()



        

```
