```python
import random
import csv
data=[]
list=[]
total=0
Fields=["List","quantity","price"]
items=["Apple","Bananas","Milk","Eggs","Bread","Pulses","Mangoes","Gauava","Kiwi","Coconut"]
for i in range(100):
    item=random.choice(items)
    price=random.randint(1,100)
    quantity=random.randint(1,10)
    list.append([item,price,quantity])
with open("23122118.csv","w",newline='\n') as file:
    csvwriter=csv.writer(file)
    csvwriter.writerow(Fields)
    csvwriter.writerows(list)
total_cost_by_item = {}
min_prices = {}
max_prices = {}
with open("23122118.csv", "r") as file:
    csvreader = csv.reader(file)
    next(csvreader)  
    for row in csvreader:
        item, quantity, price = row
        quantity = int(quantity)
        price = int(price)
        total_cost = quantity * price
        if csvreader.line_num <= 5 or csvreader.line_num >= 91:
            print(f"Item: {item}, Quantity: {quantity}, Price: {price}, Total Cost: {total_cost}")
        if item not in total_cost_by_item:
            total_cost_by_item[item] = 0
        total_cost_by_item[item] += total_cost
        if item not in min_prices or price < min_prices[item]:
            min_prices[item] = price
        if item not in max_prices or price > max_prices[item]:
            max_prices[item] = price
print("\nTotal Cost by Item:")
print("Item Name | Total Quantity | Total")
for item, total_cost in total_cost_by_item.items():
    total_quantity = sum([row[1] for row in data if row[0] == item])
    print(f"{item} | {total_quantity} | {total_cost}")
print("\nMinimum and Maximum Prices by Item:")
print("Item Name | Min Price | Max Price")
for item in items:
    min_price = min_prices.get(item, 0)
    max_price = max_prices.get(item, 0)
    print(f"{item} | {min_price} | {max_price}")
    
        

    
        

        
    


```
